```python
import os
from PIL import Image
import numpy as np
def convert_images_to_jpg(root_dir):
    # Walk through the root directory
    for subdir, dirs, files in os.walk(root_dir):
        for file in files:
            file_path = os.path.join(subdir, file)

            # Open an image file
            try:
                with Image.open(file_path).convert('RGB') as img:
                      # Create the output directory if it does not exist
                      relative_path = os.path.relpath(subdir, root_dir)
                      output_dir = os.path.join(root_dir, 'converted', relative_path)
                      os.makedirs(output_dir, exist_ok=True)
                      img2 = img.quantize(128).convert('RGB')
                      img3 = np.asarray(img, dtype='float32') - np.asarray(img2, dtype='float32')
                      img3 = (img3 - np.min(img3))
                      img3 = np.clip(img3, 0.0, 255.0)
                      img3 = np.asarray(img3, dtype='uint8')
                      img3 = Image.fromarray(img3)
                      # Save the image in JPEG format
                      file_name, _ = os.path.splitext(file)
                      new_file_path = os.path.join(output_dir, f"{file_name}.jpg")
                      img3.save(new_file_path, 'JPEG', quality=100)
                      print(f"Converted {file_path} to {new_file_path}")
            except Exception as e:
                print(f"Could not convert {file_path}: {e}")
```

```python

# Usage
dataset_dir = '/content/real_gen_dataset'  # Replace with your dataset directory path
convert_images_to_jpg(dataset_dir)
```
